## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df["line_subtotal"]= df["unit_price"] * df["qty"]
df[["unit_price","qty","line_subtotal"]].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
# quantity * weight * shipping price 
# if weight <= 50 then shipping price = 10 else spipping price = 7
df["shipping_price"] = df.apply(
    lambda row: row["qty"] * row["unit_weight"] * (7 if row["qty"] * row["unit_weight"] > 50 else 10),
    axis=1
)

# Display the first few rows to verify
df[["qty", "unit_weight", "shipping_price"]].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
#logic: total = subtotal *1.095 + shipping 
df["total_price"] = df["line_subtotal"] * 1.0925 + df["shipping_price"]

# Display the first few rows of the new column to verify
df[["line_subtotal", "shipping_price", "total_price"]].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
#logic: net_cost = cost*quantity + shipping price  

# Calculate the line cost using unit cost, quantity, and shipping price
df["line_cost"] = (df["unit_cost"] * df["qty"]) + df["shipping_price"]

# Display the first few rows of the new column to verify
df[["unit_cost", "qty", "shipping_price", "line_cost"]].head()

```

```python
# Create a column for the profit of each line using line cost and line price

# profit for each line 


df["line_profit"] = df["total_price"] - df["line_cost"]

# Display 
df[["total_price", "line_cost", "line_profit"]].head()


```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.


top_dogs = df["client_id"].value_counts().head(5).index.tolist()

top_clients_df = df[df["client_id"].isin(top_dogs)]

top_clients_spending = top_clients_df.groupby('client_id')['total_price'].sum()

print(top_clients_spending)

```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
# Group by 'client_id' and sum the 'qty' to find the top 5 clients by total quantity
client_totals = df.groupby('client_id')['qty'].sum()

#  Sort the clients by total quantity in descending order and select the top 5 clients
top_5_clients = client_totals.sort_values(ascending=False).head(5).index.tolist()

#  Filter the DataFrame to include only the top 5 clients
top_clients_df = df[df['client_id'].isin(top_5_clients)]

# Group by 'client_id' and calculate the totals
summary_df = top_clients_df.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('total_price', 'sum'),
    total_profit=('line_profit', 'sum')
).reset_index()

#  Display the summary DataFrame
print(summary_df)



```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 


# Define a function that converts a dollar amount to millions.


# Apply the currency_format_millions function to only the money columns. 


# Rename the columns to reflect the change in the money format. 

# Define the money columns
money_columns = ['total_shipping_price', 'total_revenue', 'total_profit']

# Define a function that converts a dollar amount to millions
def currency_format_millions(x):
    return x / 1_000_000

# Apply the function to the money columns
if all(col in summary_df.columns for col in money_columns):
    summary_df[money_columns] = summary_df[money_columns].applymap(currency_format_millions)
else:
    print("Some columns are missing. Check column names.")

# Rename the columns to reflect the change in money format
summary_df = summary_df.rename(columns={
    'total_shipping_price': 'Total Shipping Price (in millions)',
    'total_revenue': 'Total Revenue (in millions)',
    'total_profit': 'Total Profit (in millions)',
    'total_units_purchased': 'Total Units Purchased'
})

# Display the formatted DataFrame
print(summary_df)
```

```python
## Sort by 'Total Profit (in millions)' from highest to lowest
sorted_summary_df = summary_df.sort_values('Total Profit (in millions)', ascending=False)

# Check the sorted DataFrame
print(sorted_summary_df)

```
